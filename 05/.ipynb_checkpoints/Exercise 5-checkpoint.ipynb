First, we import all the necessary libraries we will be using. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

#### General Functions:

Here, we define all the main functions we will be using for both datasets. We created a function to split our data into training and testing set, one to normalize our data, another one to predict our dependent variable, calculate the RMSE, create the mini batches, learn using ridge regression with the mini batches and the last one to perform the k-fold cross validation. 

```python
def split_train_test(x_data,y_data,train_percentage):
    row_selection = np.random.rand(len(x_data)) < train_percentage
    x_train = x_data[row_selection]
    y_train = y_data[row_selection]
    x_test = x_data[~row_selection]
    y_test = y_data[~row_selection]
    return (x_train,y_train,x_test,y_test)

def normalize_dataframe(dataframe):
    new_dataframe = ((dataframe - dataframe.min())/(dataframe.max() - dataframe.min()))*1
    return new_dataframe

def predict(X,beta):
    return np.dot(X,beta)

def calculate_rmse(X,y,beta):
    return np.sqrt(np.square(np.subtract(y,np.dot(X,beta))).mean())

def create_mini_batches(X, y, batch_size): 
    mini_batches = [] 
    data = np.hstack((X, y))
    np.random.shuffle(data) 
    num_batches = data.shape[0] // batch_size
    for i in range(num_batches): 
        mini_batch = data[i * batch_size:(i + 1)*batch_size, :]
        X_batch = mini_batch[:, :-1] 
        y_batch = mini_batch[:, -1].reshape((-1, 1))
        mini_batches.append((X_batch, y_batch)) 
    if data.shape[0] % batch_size != 0: 
        mini_batch = data[(i + 1) * batch_size:data.shape[0]] 
        X_batch = mini_batch[:, :-1] 
        y_batch = mini_batch[:, -1].reshape((-1, 1))
        mini_batches.append((X_batch, y_batch)) 
    return mini_batches

def learn_ridgereg_mini_bgd(x_train,y_train,x_test,y_test,step_size,reg_constant,
                            batch_size,max_iterations):
    rmse_error_train = []
    rmse_error_test = []
    beta = np.zeros((x_train.shape[1],1))
    for i in range(max_iterations):
        mini_batches = create_mini_batches(x_train,y_train,batch_size)
        for batches in mini_batches:
            X_batch,y_batch = batches
            l2_loss = -2*np.dot(X_batch.T,y_batch-np.dot(X_batch,beta))
            l2_reg = 2*reg_constant*beta
            beta -= step_size*(l2_loss+l2_reg)
        rmse_train = calculate_rmse(x_train,y_train,beta)
        rmse_error_train.append(rmse_train)
        rmse_test = calculate_rmse(x_test,y_test,beta)
        rmse_error_test.append(rmse_test)
    return (beta,rmse_error_train,rmse_error_test)

def k_fold_cross_validation(X,y,k,step_size_arr,reg_constant_arr,batch_size,max_iterations):
    parameters_df = pd.DataFrame(columns=['Step Size','Reg. Constant','Avg. RMSE'])
    fold_size = len(X) // k
    folds = create_mini_batches(X,y,fold_size)
    for stp in step_size_arr:
        step_size = stp
        for con in reg_constant_arr:
            reg_constant = con
            fold_errors = []
            for i in range(len(folds)-1):
                x_test = np.array(folds[i][0])
                y_test = np.array(folds[i][1])
                x_train = []
                y_train = []
                for n in [x for x in range(len(folds)) if x != i]:
                    x_train.extend(folds[n][0])
                    y_train.extend(folds[n][1])
                x_train = np.array(x_train)
                y_train = np.array(y_train)
                coefficients,train_error,test_error = learn_ridgereg_mini_bgd(x_train,y_train,
                                                                x_test,y_test,step_size,reg_constant,
                                                                batch_size,max_iterations)
                fold_errors.append(train_error)
            error_mean = np.mean(fold_errors)
            parameters_df = parameters_df.append({'Step Size':step_size, 
                                                  'Reg. Constant':reg_constant,
                                                  'Avg. RMSE':error_mean},ignore_index=True)
    
    optimal_row = parameters_df[parameters_df['Avg. RMSE'] == np.min(parameters_df['Avg. RMSE'])]
    optimal_step_size = optimal_row['Step Size'].values[0]
    optimal_reg_constant = optimal_row['Reg. Constant'].values[0]
    lowest_rmse = optimal_row['Avg. RMSE'].values[0]
    return (parameters_df,optimal_step_size,optimal_reg_constant,lowest_rmse)
```

## Bank Marketing

### 1. Data Pre-Processing

We start by reading our bank marketing data and loading it into a pandas dataframe. We then use the .info() function to see and understand how our data is structured. 

```python
bank_data = pd.read_csv('bank.csv',sep=';')
bank_data.info()
```

After we see that we have several non-numerical columns, we create dummy variables for all of them, and dropping the first column to prevent multi-collinearity. We use as well the .info() function to see how our data changed.

```python
bank_data = pd.get_dummies(bank_data,drop_first=True)
bank_data.info()
```

We divide our data into our independent and dependent variables, and also into our training and test set. 

```python
x_data_bank = bank_data.drop('y_yes',axis=1)
y_data_bank = np.array(bank_data['y_yes']).reshape(len(bank_data),1)

x_train_bank,y_train_bank,x_test_bank,y_test_bank = split_train_test(x_data_bank,y_data_bank,0.80)
```

Once we have divided our data, we normalize our independent variables.

```python
x_train_bank = normalize_dataframe(x_train_bank)
x_test_bank = normalize_dataframe(x_test_bank)
```

### 2. Linear Classification with Gradient Descent

We train our data using a step size of 0.00001 and a regularization constant of 0.1 for 100 iterations with our ridge regression algorithm. Then we plot the RMSE on our training and testing data, where we can see that the error is minimizing after each iteration.

```python
step_size = 0.00001
reg_constant = 0.1
batch_size = 50
max_iterations = 100

coefficients,train_error,test_error = learn_ridgereg_mini_bgd(x_train_bank,y_train_bank,x_test_bank,
                                                              y_test_bank,step_size,reg_constant,
                                                              batch_size,max_iterations)

plt.figure(figsize=[12,5])

plt.subplot(1,2,1)
plt.plot(train_error)
plt.title('RMSE on Train Data')
plt.xlabel('Iterations')
plt.ylabel('RMSE')

plt.subplot(1,2,2)
plt.plot(test_error)
plt.title('RMSE on Test Data')
plt.xlabel('Iterations')
plt.ylabel('RMSE')

plt.tight_layout()
```

We train our data using a step size of 0.0001 and a regularization constant of 1 for 100 iterations with our ridge regression algorithm. Then we plot the RMSE on our training and testing data, where we can see that the error is minimizing after each iteration, although we can see that the error doesn't minimize as smoothly on our test set. 

```python
step_size = 0.0001
reg_constant = 1
batch_size = 50
max_iterations = 100

coefficients,train_error,test_error = learn_ridgereg_mini_bgd(x_train_bank,y_train_bank,x_test_bank,
                                                              y_test_bank,step_size,reg_constant,
                                                              batch_size,max_iterations)

plt.figure(figsize=[12,5])

plt.subplot(1,2,1)
plt.plot(train_error)
plt.title('RMSE on Train Data')
plt.xlabel('Iterations')
plt.ylabel('RMSE')

plt.subplot(1,2,2)
plt.plot(test_error)
plt.title('RMSE on Test Data')
plt.xlabel('Iterations')
plt.ylabel('RMSE')

plt.tight_layout()
```

We train our data using a step size of 0.001 and a regularization constant of 10 for 100 iterations with our ridge regression algorithm. Then we plot the RMSE on our training and testing data, where we can see that the error doesn't minimize after each iteration, but keeps jumping during all the iterations.

```python
step_size = 0.001
reg_constant = 10
batch_size = 50
max_iterations = 100

coefficients,train_error,test_error = learn_ridgereg_mini_bgd(x_train_bank,y_train_bank,x_test_bank,
                                                              y_test_bank,step_size,reg_constant,
                                                              batch_size,max_iterations)

plt.figure(figsize=[12,5])

plt.subplot(1,2,1)
plt.plot(train_error)
plt.title('RMSE on Train Data')
plt.xlabel('Iterations')
plt.ylabel('RMSE')

plt.subplot(1,2,2)
plt.plot(test_error)
plt.title('RMSE on Test Data')
plt.xlabel('Iterations')
plt.ylabel('RMSE')

plt.tight_layout()
```

### 3. Hyper-Parameter Tuning

Here we define our grid of size 3x3 for our grid search, and we then run our k-fold cross validation algorithm to find the best parameters for our model. For each combination of parameters, we print the average RMSE, and we then print the optimal combination with its average RMSE. 

```python
k = 5
step_size_arr = [0.00001,0.0001,0.001]
reg_constant_arr = [0.1,1,10]
batch_size = 50
max_iterations = 100

parameters_df,optimal_step_size,optimal_reg_constant,lowest_rmse = k_fold_cross_validation(
    x_train_bank,y_train_bank,k,step_size_arr,reg_constant_arr,batch_size,max_iterations)

print(parameters_df)
print('\nLowest RMSE: {}\nOptimal Step Size: {}\nOptimal Reg. Constant: {}'.
      format(lowest_rmse,optimal_step_size,optimal_reg_constant))
```

We show a 3D scatter plot of our results using the three axes we have, step size, regularization constant and average RMSE.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(parameters_df['Step Size'], parameters_df['Reg. Constant'], parameters_df['Avg. RMSE'], 
           marker='o')
ax.set_xlabel('Step Size')
ax.set_ylabel('Reg. Constant')
ax.set_zlabel('Avg. RMSE')
```

Once we have found our optimal parameters, we run the ridge regression learning algorithm again with these parameters. We then plot the RMSE on the training and testing set, where we can see that the error is minimizing after each iteration.

```python
step_size = optimal_step_size
reg_constant = optimal_reg_constant
batch_size = 50
max_iterations = 100

coefficients,train_error,test_error = learn_ridgereg_mini_bgd(x_train_bank,y_train_bank,x_test_bank,
                                                              y_test_bank,step_size,reg_constant,
                                                              batch_size,max_iterations)

plt.figure(figsize=[12,5])

plt.subplot(1,2,1)
plt.plot(train_error)
plt.title('RMSE on Train Data')
plt.xlabel('Iterations')
plt.ylabel('RMSE')

plt.subplot(1,2,2)
plt.plot(test_error)
plt.title('RMSE on Test Data')
plt.xlabel('Iterations')
plt.ylabel('RMSE')

plt.tight_layout()
```

**Comparison:**
In the graph below, we can see the model that peformed better, when we didn't implement the grid search, which had a step size of 0.0001 and a regular constant of 1. We can see that the k-fold cross validation did a great job choosing the best parameters, which were a step size of 0.001 and a regular constant of 0.1. Although the graph is not as smooth with the best parameters, it did a better job minimizing the error and it did it faster as well.
